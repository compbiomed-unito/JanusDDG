```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.metrics import root_mean_squared_error
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
#LIST OF PDB INFO FOR EACH PROTEIN

#N.B. Don't print the entire list.

with open("proteins_dict_esm2.pkl", "rb") as tf:
    proteins_dict = pickle.load(tf)

#es:
proteins_dict['1a0fA']

#N.B. Don't print the entire list.
```

```python
df_doppie = pd.read_csv('../results/Result_Ptmul_D.csv')#('../git_JANUS_DDG/Results/Result_dataset_doppie_ludovica.csv')
df_doppie = df_doppie[df_doppie['MTS'].map(lambda x: len(x.split('_'))==2)]
df_doppie['DDG']=df_doppie['DDG']

def distance3d(proteins_dict, id, MTS):
    pos1 = int(MTS.split('_')[0][1:-1]) -1 
    pos2 = int(MTS.split('_')[1][1:-1]) -1
    d_3d = proteins_dict[id[:-1].lower()+id[-1]]['distance_matrix'][pos1,pos2]
    return d_3d

df_doppie['Distance_3D'] = df_doppie.apply(lambda x: distance3d(proteins_dict, x['ID'],x['MTS']), axis = 1)
df_doppie['Distance_in_seq'] = df_doppie['MTS'].map(lambda x: (np.abs(int(x.split('_')[0][1:-1]) - int(x.split('_')[1][1:-1]))))
```

```python
distance_list_3d = []
absolute_err = []

for i in range(df_doppie.shape[0]):
    df_doppie_window = df_doppie.iloc[i]

    if len(df_doppie_window) > 1:
        absolute_err.append(np.abs((np.array(df_doppie_window['DDG']) - np.array(df_doppie_window['DDG_JanusDDG']))))

    else:
        absolute_err.append(np.nan)

    distance_list_3d.append(df_doppie_window['Distance_3D'])

rmse_df_3d = pd.Series(absolute_err)
```

```python
distance_list = []
absolute_err = []

for i in range(df_doppie.shape[0]):
    df_doppie_window_i = df_doppie.iloc[i]

    if len(df_doppie_window_i) > 1:
        absolute_err.append(np.abs(np.array(df_doppie_window_i['DDG']) - np.array(df_doppie_window_i['DDG_JanusDDG'])))

    else:
        absolute_err.append(np.nan)

    distance_list.append(df_doppie_window_i['Distance_in_seq'])

rmse_df_seq = pd.Series(absolute_err)
```

```python
rmse_df_seq
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.scatterplot(ax=axes[0], data=df_doppie, x='DDG', y='DDG_JanusDDG', hue='Distance_in_seq', palette='coolwarm')
x_vals = np.linspace(df_doppie['DDG'].min(), df_doppie['DDG'].max(), 300)
axes[0].plot(x_vals, x_vals, linestyle='dashed', color='black', linewidth=1.2)

axes[0].set_xlabel("DDG", fontsize=12)
axes[0].set_ylabel("JanusDDG Predictions", fontsize=12)
axes[0].set_title("Scatter plot of predictions", fontsize=14, fontweight = 'bold')
axes[0].legend(title= "Seq. Separation", bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0].grid(True, linestyle= '--', alpha=0.6)

smoothed = lowess(rmse_df_seq, distance_list, frac=0.4)  # frac controlla il livello di smoothing
sns.scatterplot(ax=axes[1], x=distance_list, y=rmse_df_seq, color='red', s=50, label="Original Points")
sns.lineplot(ax=axes[1], x=smoothed[:, 0], y=smoothed[:, 1], color='blue', linewidth=2, label="Smoothed Correlation")

axes[1].set_xlabel("Sequence Spearation", fontsize=12)
axes[1].set_ylabel("Absolute Error", fontsize=12)
axes[1].set_title("Absolute Error", fontsize=14, fontweight = 'bold')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Primo subplot - scatter plot invariato
sns.scatterplot(ax=axes[0], data=df_doppie, x='DDG', y='DDG_JanusDDG', hue='Distance_in_seq', palette='coolwarm')
x_vals = np.linspace(df_doppie['DDG'].min(), df_doppie['DDG'].max(), 300)
axes[0].plot(x_vals, x_vals, linestyle='dashed', color='black', linewidth=1.2)
axes[0].set_xlabel("DDG", fontsize=12)
axes[0].set_ylabel("JanusDDG Predictions", fontsize=12)
axes[0].set_title("Scatter plot of predictions", fontsize=14, fontweight='bold')
axes[0].legend(title="Seq. Separation", bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0].grid(True, linestyle='--', alpha=0.6)

# Secondo subplot - con banda di confidenza locale
smoothed = lowess(rmse_df_seq, distance_list, frac=0.4)  # frac controlla il livello di smoothing

# Calcola l'intervallo di confidenza locale considerando il numero di punti
def calculate_confidence_interval(x_smooth, y_smooth, x_orig, y_orig, window_size=0.3, confidence=0.95):
    """Calcola l'intervallo di confidenza locale per ogni punto della curva smoothed"""
    from scipy import stats
    
    confidence_errors = []
    x_range = np.max(x_orig) - np.min(x_orig)
    
    for i, (x_s, y_s) in enumerate(zip(x_smooth, y_smooth)):
        # Definisci una finestra locale attorno al punto
        window = window_size * x_range
        mask = np.abs(x_orig - x_s) <= window
        n_points = np.sum(mask)
        
        if n_points > 2:  # Servono almeno 3 punti per calcolare statistiche affidabili
            local_residuals = y_orig[mask] - np.interp(x_orig[mask], x_smooth, y_smooth)
            local_std = np.std(local_residuals, ddof=1)  # Sample standard deviation
            
            # Calcola l'intervallo di confidenza usando la distribuzione t
            t_value = stats.t.ppf((1 + confidence) / 2, df=n_points-1)
            margin_of_error = t_value * local_std / np.sqrt(n_points)
            confidence_errors.append(margin_of_error)
        else:
            # Se abbiamo pochi punti, usa un errore pi grande (penalit per incertezza)
            global_std = np.std(y_orig)
            confidence_errors.append(global_std * 2)  # Errore conservativo
    
    return np.array(confidence_errors)

# Calcola l'intervallo di confidenza locale
confidence_errors = calculate_confidence_interval(smoothed[:, 0], smoothed[:, 1], 
                                                np.array(distance_list), np.array(rmse_df_seq))

# Plot dei punti originali
sns.scatterplot(ax=axes[1], x=distance_list, y=rmse_df_seq, color='red', s=50, label="Original Points")

# Plot della linea smoothed
sns.lineplot(ax=axes[1], x=smoothed[:, 0], y=smoothed[:, 1], color='blue', linewidth=2, label="Smoothed Line")

# Aggiungi banda di confidenza al 95%
axes[1].fill_between(smoothed[:, 0], 
                     smoothed[:, 1] - confidence_errors, 
                     smoothed[:, 1] + confidence_errors, 
                     alpha=0.3, color='blue', label='95% Confidence Band')

axes[1].set_xlabel("Sequence Separation", fontsize=12)
axes[1].set_ylabel("Absolute Error", fontsize=12)
axes[1].set_title("Absolute Error", fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
